# Dropping der Features

- Filtern des Datensatzes auf Basis .md-File 08_Feature_Selection


```python
import os
import pandas as pd
import numpy as np
```

```python

pickle_path = '../../data/processed/23_cleaned_master_data.pkl'

df = pd.read_pickle(pickle_path)

columns_to_drop = [
    'time_gap', 'time_gap_seconds', 'avg_speed',  # Harte Data Leakage Spalten (wissen wir erst im Ziel)
    'birthdate',                                  # Komplett redundant zu 'age_at_race'
    'race_ranking',                               # Haben wir durch den 'race_competitiveness_median' ersetzt
    'wind_speed_delta', 'wind_direction_delta',   # Roh-Zwischenschritte (sind im 'wind_stability_index' drin)

    # Die 12 alten Wetter-Rohspalten (da wir jetzt eure Hybrid-Features & den Wind-Index nutzen)
    'departure_temp_mittel', 'departure_regen_mittel', 'departure_wind_mittel',
    'departure_luftfeuchte_mittel', 'departure_niederschlag_mittel', 'departure_windrichtung_mittel',
    'arrival_temp_mittel', 'arrival_regen_mittel', 'arrival_wind_mittel',
    'arrival_luftfeuchte_mittel', 'arrival_niederschlag_mittel', 'arrival_windrichtung_mittel'
]

# Spalten droppen (errors='ignore' sorgt dafür, dass es nicht abstürzt, falls eine Spalte schon fehlt)
df_final = df.drop(columns=columns_to_drop, errors='ignore')

print(f"Spalten erfolgreich eliminiert!")
print(f"   -> Dimensionen der Feature-Matrix (X): {df_final.shape} (Zeilen, Spalten)")

# Kontrolle

for i, col in enumerate(df_final.columns, 1):
    print(f" {i:02d}. {col}")
```

```python
# wir kennzeichnen die MetaDaten Spalten noch mit meta_

metadata_rename_dict = {
    'race': 'meta_race',
    'year': 'meta_year',
    'url': 'meta_url',
    'rider_url': 'meta_rider_url',
    'name': 'meta_name',
    'nationality': 'meta_nationality',
    'url_name': 'meta_url_name',
    'departure': 'meta_departure',
    'arrival': 'meta_arrival',
    'date': 'meta_date',
    'current_team': 'meta_current_team',
    'departure_lat': 'meta_departure_lat',
    'departure_lon': 'meta_departure_lon',
    'arrival_lat': 'meta_arrival_lat',
    'arrival_lon': 'meta_arrival_lon'
}

# Spalten umbenennen
df_final.rename(columns=metadata_rename_dict, inplace=True)

for i, col in enumerate(df_final.columns, 1):
    print(f" {i:02d}. {col}")
```

```python
# Checkpoint
pfad = '../../data/processed/24_cleaned_master_data.pkl'
df_final.to_pickle(pfad)
```
