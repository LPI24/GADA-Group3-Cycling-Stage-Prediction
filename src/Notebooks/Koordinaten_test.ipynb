```python
import pandas as pd
import numpy as np

# Pickle-Datei einlesen
df0 = pd.read_pickle("/Users/leonfrey/Documents/TU Dresden/10 Semester/Gestaltungsansätze der Advanced Data Analytics/4_Version/cleaned_master_data_coordinates_fixed_3.pkl")

# Haversine-Funktion zur Berechnung der Entfernung in Kilometern
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Erdradius in km

    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))

    return R * c

# Entfernung berechnen und als neue Spalte hinzufügen
df0["distance_km_koordinaten"] = haversine(
    df0["departure_lat"],
    df0["departure_lon"],
    df0["arrival_lat"],
    df0["arrival_lon"]
)

# Ergebnis als Pickle-Datei speichern
df0.to_pickle("/Users/leonfrey/Documents/TU Dresden/10 Semester/Gestaltungsansätze der Advanced Data Analytics/4_Version/cleaned_master_data_coordinates_fixed_3_Koordinatendistanz.pkl")

print("Fertig! Datei gespeichert.")
```

```python
import pickle
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from geopy.distance import geodesic
from geopy.point import Point


# --- 1. Pickle laden ---
with open('/Users/leonfrey/Documents/TU Dresden/10 Semester/Gestaltungsansätze der Advanced Data Analytics/4_Version/cleaned_master_data_coordinates_fixed_3_Koordinatendistanz_ohneKM_157rows.pkl', "rb") as f:
    df = pickle.load(f)


# --- 2. Geocoding einrichten ---
geolocator = Nominatim(user_agent="mein_wetter_projekt_v2", timeout=10)

# ✅ min_delay_seconds=2: mind. 2s zwischen Anfragen (Nominatim-Policy: max 1 req/s)
# ✅ max_retries=3: bei Fehler 3x erneut versuchen
# ✅ error_wait_seconds=5: bei Fehler (z.B. 429) 5s warten vor Retry
geocode = RateLimiter(
    geolocator.geocode,
    min_delay_seconds=3,
    max_retries=3,
    error_wait_seconds=15,
    swallow_exceptions=False
)

RADIUS_KM = 264


def get_coords(ortsname):
    try:
        loc = geocode(str(ortsname))
        if loc:
            return loc.latitude, loc.longitude
    except Exception:
        pass
    return None, None


def get_coords_near_departure(arrival_name, dep_lat, dep_lon):
    if dep_lat is None or dep_lon is None:
        print(f"  ⚠ Keine departure-Koordinaten – Suche für '{arrival_name}' übersprungen.")
        return None, None

    delta = RADIUS_KM / 111.0

    viewbox = [
        Point(dep_lat + delta, dep_lon - delta),  # Nord-West
        Point(dep_lat - delta, dep_lon + delta),  # Süd-Ost
    ]

    try:
        results = geocode(
            str(arrival_name),
            exactly_one=False,
            viewbox=viewbox,
            bounded=True,
            limit=5
        )
    except Exception as e:
        print(f"  ⚠ Geocoding-Fehler für '{arrival_name}': {e}")
        return None, None

    if not results:
        print(f"  ✗ '{arrival_name}' nicht im Umkreis von {RADIUS_KM} km um "
              f"({dep_lat:.4f}, {dep_lon:.4f}) gefunden.")
        return None, None

    for result in results:
        dist = geodesic((dep_lat, dep_lon), (result.latitude, result.longitude)).km
        if dist <= RADIUS_KM:
            return result.latitude, result.longitude

    closest_dist = geodesic(
        (dep_lat, dep_lon),
        (results[0].latitude, results[0].longitude)
    ).km
    print(f"  ✗ '{arrival_name}' – Treffer außerhalb {RADIUS_KM} km "
          f"(nächster: {closest_dist:.1f} km).")
    return None, None


# --- 3. Koordinaten für departure ermitteln (Cache) ---
departure_orte = df["departure"].dropna().unique()
dep_coords_cache = {}
for ort in departure_orte:
    dep_coords_cache[ort] = get_coords(ort)
    print(f"Departure '{ort}': {dep_coords_cache[ort]}")

df["departure_lat"] = df["departure"].map(lambda o: dep_coords_cache.get(o, (None, None))[0])
df["departure_lon"] = df["departure"].map(lambda o: dep_coords_cache.get(o, (None, None))[1])


# --- 4. Arrival-Koordinaten: nur im 264-km-Radius um departure ---
arrival_cache = {}

for _, row in df.iterrows():
    key = (row["departure"], row["arrival"])
    if key in arrival_cache or pd.isna(row["arrival"]):
        continue

    dep_lat = dep_coords_cache.get(row["departure"], (None, None))[0]
    dep_lon = dep_coords_cache.get(row["departure"], (None, None))[1]

    arr_lat, arr_lon = get_coords_near_departure(row["arrival"], dep_lat, dep_lon)
    arrival_cache[key] = (arr_lat, arr_lon)

df["arrival_lat"] = df.apply(
    lambda r: arrival_cache.get((r["departure"], r["arrival"]), (None, None))[0], axis=1
)
df["arrival_lon"] = df.apply(
    lambda r: arrival_cache.get((r["departure"], r["arrival"]), (None, None))[1], axis=1
)


# --- 5. Speichern ---
df.to_pickle('/Users/leonfrey/Documents/TU Dresden/10 Semester/Gestaltungsansätze der Advanced Data Analytics/4_Version/cleaned_master_data_coordinates_fixed_3_Koordinatendistanz_ohneKM_157rows_richtigerArrival.pkl')

print(df[["departure", "departure_lat", "departure_lon",
          "arrival",   "arrival_lat",   "arrival_lon"]].head())
```
