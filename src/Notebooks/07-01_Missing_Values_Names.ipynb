```python
# 1. Wir erstellen eine Teilmenge, die nur die Zeilen mit fehlenden Namen enthält
df_nas = df[df['name'].isna()]

print("=== 1. JAHRE MIT DEN MEISTEN NA-WERTEN ===")
display(df_nas['year'].value_counts().head(20))

print("\n=== 2. RENNEN MIT DEN MEISTEN NA-WERTEN ===")
display(df_nas['race'].value_counts().head(20))

print("\n=== 3. PLATZIERUNGEN (RANK) MIT DEN MEISTEN NA-WERTEN ===")
# Hier sehen wir, ob z.B. bestimmte Platzierungen (wie 'DNF' oder 'DNS') oft keinen Namen haben
display(df_nas['rank'].value_counts().head(20))

print("\n=== 4. KOMBINATION: WELCHE JAHRE + RENNEN SIND AM STÄRKSTEN BETROFFEN? ===")
# Gruppierung nach Jahr und Rennen, um systematische Fehler (z.B. ein komplett fehlendes Jahr bei einem Rennen) zu finden
kombi_stats = df_nas.groupby(['year', 'race', 'stage_nr']).size().reset_index(name='Anzahl_Missing_Names')
kombi_stats = kombi_stats.sort_values(by='Anzahl_Missing_Names', ascending=False)

# Anzeige der Top 20 betroffenen Jahr-Rennen-Kombinationen
display(kombi_stats.head(20))
```
